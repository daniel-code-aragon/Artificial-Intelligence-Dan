```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re
import seaborn as sns
```

```python

# Ruta relativa al archivo
ruta_archivo = os.path.join('data', 'houses_Madrid.csv')

# Obtener la ruta del directorio actual del cuaderno
directorio = os.getcwd()

# Combinar la ruta del directorio actual con la ruta relativa al archivo
ruta_absoluta = os.path.join(directorio, ruta_archivo)

madrid_df=pd.read_csv(ruta_absoluta)

madrid_df.info()
```

-  Lo primero que se va a hacer va a ser eliminar columnas que no aporten demasiada informacion para este trabajo 
,que su informacion ya este contenida en otras o que tenga una alta o total canitidad de valores nulos.


```python
madrid_df = madrid_df.dropna(axis=1, how='all')

madrid_df = madrid_df.drop(['Unnamed: 0',
                            'title',
                           'is_exact_address_hidden',
                           'street_number',
                           'is_rent_price_known',
                           'is_buy_price_known',
                           'house_type_id',
                           'is_orientation_north',
                           'is_orientation_west',
                           'is_orientation_east',
                           'is_orientation_south',
                           'sq_mt_useful'
                           ,'sq_mt_allotment',
                           'operation',
                           'is_new_development',
                           'built_year',
                           'has_garden'],
                          axis=1)

madrid_df.info()
```

```python
madrid_df.describe(include='all')
```

# Transformaciones con las diferentes columnas

```python
#valores nulos de numero de plantas sustituidos por 1 asumiendo que si esta vacio solo tiene una planta
#valores nulos sustituidos por False
madrid_df['n_floors']=madrid_df['n_floors'].fillna(1)
madrid_df['is_floor_under']=madrid_df['is_floor_under'].fillna(False)
#madrid_df.head()
```

```python
#Extraigo de la columna 'neigborhood_id' mediante regular expresions cual es el distrito de madrid donde esta la vivienda y
# creo una nueva columna con esos datos

def obtener_distrito(texto):
    patron = r"District\s+\d+:\s+(.*)"
    resultado = re.search(patron, texto)
    if resultado:
        texto_despues_district = resultado.group(1)
        #print(texto_despues_district)
    

    return texto_despues_district


madrid_df['Distrito'] = madrid_df['neighborhood_id'].apply(obtener_distrito)
```

```python
madrid_df.head(5)
```

```python
#hago una transformacion a las columnas que especifican una caracteritica concreta de la vivienda, 
# Asumindo que si el campo esta vacio es que no posee esa caracteristica

columnas_caracteristicas=['has_central_heating','has_individual_heating','has_ac',
                         'has_fitted_wardrobes','has_lift','is_exterior','has_pool','has_terrace','has_balcony','has_storage_room',
                         'is_accessible','has_green_zones','has_parking','is_parking_included_in_price']

madrid_df[columnas_caracteristicas]=madrid_df[columnas_caracteristicas].fillna(False)

madrid_df['n_bathrooms']=madrid_df['n_bathrooms'].fillna(2)  # Numero entero mas cercano a la media 1.83
```

```python
madrid_df.info()
```

```python
# de los las aproximadas 22k entradass 2k no especifica en piso por lo que elimino esas entradas
madrid_df = madrid_df.dropna(subset='floor')
```

```python
# Asi tengo valores numericos que son mas faciles de utilizar a la hora de visualizar 
madrid_df = madrid_df.replace({True: 1, False: 0}) 
```

```python
# Quiero ver los valores que se le estan dando a la columna'energy_certificate'
madrid_df['energy_certificate'].value_counts()
```

```python
#Asigno una puntuacion del 0 al 6 en funcion del certificado energitico para darle un valor numerico

madrid_df['puntuacion_energetica'] = madrid_df['energy_certificate'].replace({'en trámite': 0,
                                                                           'no indicado': 0,
                                                                          'inmueble exento':0,
                                                                              'G':1,
                                                                             'F':2,
                                                                              'E':3,
                                                                              'D':4,
                                                                              'C':4,
                                                                              'B':5,
                                                                              'A':6
                                                                             })
madrid_df['puntuacion_energetica'].value_counts()
```

```python
precios = madrid_df['buy_price']



plt.figure(figsize=(8, 4))
plt.hist(precios, bins=10, color='blue', edgecolor='black')

plt.title('Distribución de precios de las casas')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')

plt.show()
```

```python

distritos = madrid_df['Distrito']

distritos_count = distritos.value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=distritos_count.index, y=distritos_count.values, color='blue')

plt.title('Distribución de casas por zona')
plt.xlabel('Distrito')
plt.ylabel('Cantidad de pisos ofertados')


plt.xticks(rotation=45)


plt.show()
```

```python
# Visualizacion de variables numericas con un grafico de dispersión.
#la que mas afecta son los metros cuadrados

precio=madrid_df['buy_price']
m2=madrid_df['sq_mt_built']

plt.figure(figsize=(12, 5))
plt.scatter(m2, precio, color='blue', alpha=0.5)

plt.title('Relación entre m2s y el precio')
plt.xlabel('m2')
plt.ylabel('Precio')


plt.show()
```

```python
categoria = madrid_df[madrid_df['puntuacion_energetica'] != 0]['puntuacion_energetica']
precio = madrid_df['buy_price']

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=categoria, y=precio)

# Personalizar el gráfico
plt.title('Relación entre Categoría y Precio')
plt.xlabel('Categoría')
plt.ylabel('Precio')

# Mostrar el gráfico
plt.show()
```

```python
variables_bool = ['has_individual_heating', 'has_ac', 'has_fitted_wardrobes',
                        'has_lift','is_exterior','has_pool','has_terrace','has_balcony','has_storage_room',
                        'is_accessible','has_green_zones','has_parking']

precio = madrid_df['buy_price']

fig, axs = plt.subplots(len(variables_bool), figsize=(8, 50))


for i, var_cat in enumerate(variables_bool):
    sns.violinplot(x=madrid_df[var_cat], y=precio, ax=axs[i])
    axs[i].set_title('Relación entre {} y Precio'.format(var_cat))
    axs[i].set_xlabel(var_cat)
    axs[i].set_ylabel('Precio')


plt.tight_layout()

plt.show()
```

```python
#Esto muestra un poco lo que decia la columna de correlacion de cada variable frente al precio, el hecho de que la casa
#posea alguna de estas caracteristicas , influye minimamente en el precio haciendolo mas caro.
```

```python

precio_medio = madrid_df.groupby('Distrito')['buy_price'].mean()
precio_medio = precio_medio.sort_values()

precio_medio_m2 = madrid_df.groupby('Distrito')['buy_price_by_area'].mean()
precio_medio_m2 = precio_medio_m2.sort_values()
#type(precio_medio)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(22, 10))


ax1.bar(precio_medio.index, precio_medio.values)
plt.xticks(rotation=45)
ax1.set_ylabel('Precio Medio Vivienda')
ax1.set_xlabel('Distrito')
ax1.set_title('Precio Vivienda')


ax2.bar(precio_medio_m2.index, precio_medio_m2.values)
ax2.set_ylabel('Precio m2')
ax2.set_xlabel('Distrito')
ax2.set_title('Precio m2')

plt.xticks(rotation=45)
plt.show()
```

```python
# Estos dos graficosm indican como varia el precio en funcion del distrito en el que este ubicado la casa:
# Siendo los mas caros: Salamanc, Chamberí , Zona Centro , Chamartin y Retiro ( las zonas ma lujosas de Madrid).
```
