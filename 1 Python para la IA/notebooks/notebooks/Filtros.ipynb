```python
import pandas as pd
import random

# Crear un DataFrame con 10 filas y 10 columnas
data = {'Columna1': [random.randint(1, 100) for _ in range(10)],
        'Columna2': [random.randint(1, 100) for _ in range(10)],
        'Columna3': ['Categoría A', 'Categoría B', 'Categoría A', 'Categoría C', 'Categoría B',
                     'Categoría C', 'Categoría A', 'Categoría B', 'Categoría C', 'Categoría A'],
        'Columna4': ['Rojo', 'Verde', 'Azul', 'Rojo', 'Verde', 'Azul', 'Rojo', 'Verde', 'Azul', 'Rojo'],
        'Columna5': ['Sí', 'No', 'No', 'Sí', 'Sí', 'No', 'No', 'Sí', 'No', 'Sí'],
        'Columna6': [random.uniform(0, 1) for _ in range(10)],
        'Columna7': [random.choice(['Perro', 'Gato', 'Pájaro']) for _ in range(10)],
        'Columna8': [random.choice(['Manzana', 'Plátano', 'Naranja']) for _ in range(10)],
        'Columna9': [random.randint(1, 5) for _ in range(10)],
        'Columna10': [random.choice(['Alto', 'Medio', 'Bajo']) for _ in range(10)]}

df = pd.DataFrame(data)

# Visualizar el DataFrame
display(df)
```

```python
# Filtrar filas donde el valor en 'Columna1' sea mayor que 50

filtro_numerico = df[df['Columna1'] > 50]
print(filtro_numerico)
```

```python
# Filtrar filas donde 'Columna3' sea igual a 'Categoría A'
filtro_categorico = df[df['Columna3'] == 'Categoría A']
display(filtro_categorico)
```

```python
# Filtrar filas donde 'Columna1' sea mayor que 50 y 'Columna3' sea igual a 'Categoría A'
filtro_multiple = df[(df['Columna1'] > 50) & (df['Columna3'] == 'Categoría A')]
display(filtro_multiple)
```

```python
# Filtrar filas donde 'Columna1' sea mayor que 70 o 'Columna4' sea igual a 'Rojo'
filtro_or = df[(df['Columna1'] > 70) | (df['Columna4'] == 'Rojo')]
display(filtro_or)
```

```python
# Filtrar filas donde 'Columna3' esté en la lista ['Categoría A', 'Categoría B']
categorias = ['Categoría A', 'Categoría B']
filtro_isin = df[df['Columna3'].isin(categorias)]
display(filtro_isin)
```

```python
# Filtrar filas donde 'Columna6' tenga valores faltantes
filtro_nan = df[df['Columna6'].isna()]
display(filtro_nan)
```

```python
suma_filas = df.sum(axis=1)
print("Suma de filas:")
print(suma_filas)
```

```python
media_filas = df.mean(axis=1)
print("Media de filas:")
print(media_filas)
```

```python
minimo_filas = df.min(axis=1)
maximo_filas = df.max(axis=1)
print("Mínimo de filas:")
print(minimo_filas)
print("Máximo de filas:")
print(maximo_filas)
```
