```python
# Creación aleatoria de datos
import numpy as np
import matplotlib.pyplot as plt
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,
              -28, -50, 60, -100, -10, -11, 13, 19, 1, 9])
```

```python
# Medidas de tendencia central
from scipy import stats
media = np.mean(X)
mediana = np.median(X)
moda, _ = stats.mode(X, keepdims=True )
print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda[0])
```

```python
# Medidas de posición
import numpy as np
import matplotlib.pyplot as plt

Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1
lim_inf = Q1 - 1.5*RIC
lim_sup = Q3 + 1.5*RIC

print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
???

plt.boxplot(X)
plt.grid()
plt.title('Diagrama de caja y bigotes')
plt.show()
```

```python
# Medidas de dispersión
desvest = np.std(X)
Var= np.std(X)**2
print('Desviación típica: ', desvest)
print('Varianza: ', Var)
```

```python
# Medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html
print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))

if curtosis>0:
    print('Distribución Leptocúrtica')
elif curtosis<0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

```python
# Cargar el set de datos de Iris
from sklearn import datasets
from pprint import pprint
iris = datasets.load_iris()
#pprint(iris)
X = iris.data
y = iris.target
print(iris.feature_names, '\n')
# ¿Qué tipo de asimetría se observa en la variable "longitud del pétalo"?

petal_length= X[:,2]

asimetria=stats.skew(petal_length)
print('Asimetria (petal lenght: )', asimetria)

# ¿A qué tipo de distribución se ajusta la variable "ancho del sépalo", según su curtosis?

curtosis = stats.kurtosis(petal_length, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html

print('Curtosis ', np.round(curtosis, 4))

if curtosis>0:
    print('Distribución Leptocúrtica')
elif curtosis<0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

```python
# EJERCICIO EXTRA SOBRE EL DATASET DE IRIS
# Extraer la media y la desviación típica de cada variable en función de la clase y representarlo en un gráfico (plt.errorbar)

colors=['blue', 'green', 'red']

for i in range(3):
    datos_clase=X[y==i]
    media_clase=np.mean(datos_clase, axis=0)
    desv_clase= np.std(datos_clase, axis =0)
    
    col = ''.join(['o', colors[i][0]])  #MArcadores circulares del color de la linea
    #print(col)
    plt.plot(range(4), media_clase, col)
    plt.errorbar(range(4), media_clase, yerr=desv_clase, color=colors[i])
    
    
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.grid()
plt.legend(['setosa', 'versicolor', 'virginica'])
plt.show()
```
