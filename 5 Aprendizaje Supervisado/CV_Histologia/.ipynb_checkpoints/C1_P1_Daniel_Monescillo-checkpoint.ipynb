## Convocatoria 1 - Proyecto 1

#### 0) Cargar una de las imágenes histológicas

```python
# Utilizar la librería skimage.io para leer la imagen 'histo_x.jpg' en formato RGB.
# Normalizar la imagen para que los píxeles se encuentren en el rango [0, 1]
# Visualizar la imagen
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage import io , morphology , color
from skimage.measure import label, regionprops
from skimage.morphology import convex_hull_image
from tabulate import tabulate
import math
import pprint


img_1 = io.imread('histo_1.jpg') #Imagen en RGB
plt.imshow(img_1, cmap='gray')
plt.show()

rgb_1 = img_1.copy()
rgb_p1 = rgb_1.astype('uint8')/255
```

#### 1) Realizar una transformación de color para convertir la imagen al espacio de color CMYK

```python
# Extraer la componente magenta de la imagen (que corresponde a la región tisular)
# Visualizar la imagen del canal magenta

with np.errstate(invalid='ignore', divide='ignore'):
    K = 1 - np.max(rgb_p1, axis=2)
    # Extrae los canales de acuerdo con la ecuación del power point
    C = (1-rgb_p1[:,:,0] - K)/(1-K)
    M = (1-rgb_p1[:,:,1] - K)/(1-K)
    Y = (1-rgb_p1[:,:,2] - K)/(1-K)
    
CMYK = (np.dstack((C,M,Y,K))*255).astype('uint8')
C,M,Y,K = cv2.split(CMYK)

fig, ax = plt.subplots(2,3)
ax[0,0].imshow(img_1, cmap='gray'), ax[0,0].set_title('original')
ax[0,0].axis('off')
ax[0,1].imshow(CMYK.astype('uint8'), cmap='gray'), ax[0,1].set_title('CMYK')
ax[0,1].axis('off')
ax[0,2].imshow(C.astype('uint8'), cmap='gray'), ax[0,2].set_title('cyan')
ax[0,2].axis('off')
ax[1,0].imshow(M.astype('uint8'), cmap='gray'), ax[1,0].set_title('magenta')
ax[1,0].axis('off')
ax[1,1].imshow(Y.astype('uint8'), cmap='gray'), ax[1,1].set_title('yellow')
ax[1,1].axis('off')
ax[1,2].imshow(K.astype('uint8'), cmap='gray'), ax[1,2].set_title('black')
ax[1,2].axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.imshow(M, cmap='gray')
plt.show()
```

#### 2) Umbralizar la imagen para separar los píxeles del fondo de la región tisular

```python
# Aplicar un filtro gaussiano de tamaño 5x5 y después utilizar el método de Otsu de manera que
# los píxeles correspondientes al lumen y al background de la imagen sean 1s y el resto de los píxeles tengan un valor de 0.
# Nota: Recordar que el método de Otsu requiere como input una imagen en el rango [0-255] en formato "uint8".
# Visualizar la máscara resultante

filtered_M = cv2.GaussianBlur(M, (5, 5), 100)

figs, ax = plt.subplots(1,3,figsize=(8, 8))
ax[0].imshow(img_1, cmap='gray') , ax[0].set_title('Img Histología')
ax[0].axis('off')
ax[1].imshow(M, cmap='gray'), ax[1].set_title('Img M channel')
ax[1].axis('off')
ax[2].imshow(filtered_M, cmap='gray'), ax[2].set_title('Img M channel Gauss')
ax[2].axis('off')
plt.show()

if np.all((filtered_M >= 0) & (filtered_M <= 255)):
    print("El array está en el rango [0, 255]")
else:
    print("El array NO está en el rango [0, 255]")
    
plt.imshow(filtered_M, cmap='gray')
plt.show()
```

```python
plt.hist(filtered_M.ravel(), bins=100)
plt.grid(True)
plt.show()

t, mask = cv2.threshold(filtered_M,0,1,cv2.THRESH_OTSU)# 0 es por defecto y 1 es el valor máximo de la máscara
print('Valor Umbral: ',t)



binary_mask = 1 - (mask).astype(np.uint8)
print(np.unique(binary_mask))
# Visualizamos para corroborar que se obtiene el mismo resultado
plt.imshow(binary_mask, cmap='gray')
plt.title('Máscara Otsu (Binary mask) t=' + str(t))
plt.axis('off')
plt.show()
```

#### 3) Limpiar la imagen eliminando los artefactos de lumen (objetos blancos pequeños que no son lúmenes)

```python
# Utilizar la librería skimage.morphology.remove_small_objects para eliminar aquellos objetos cuya área sea menor a 300 píxeles
# Más información en https://scikit-image.org/docs/dev/api/skimage.morphology.html#skimage.morphology.remove_small_objects
# Visualizaer la máscara resultante

#gray_image = color.rgb2gray(binary_mask)
min_size = 300# Tamaño mínimo del objeto a mantener
cleaned = morphology.remove_small_objects(mask, min_size=1000)
cleaned = morphology.remove_small_holes(cleaned)
# Mostrar imágenes
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(mask, cmap='gray')
axes[0].set_title('Imagen binaria original')
axes[1].imshow(cleaned, cmap='gray')
axes[1].set_title('Imagen etiquetada después de eliminar objetos pequeños')
```

```python
cleaned_mask =1 - (cleaned).astype(np.uint8)

plt.imshow(cleaned_mask, cmap='gray')
plt.title('Máscara Otsu (Cleaned mask) t=' + str(t))
plt.axis('off')
plt.show()
```

#### 4) Rellenar con 0s el fondo de la imagen para quedarnos únicamente con los lúmenes

```python
# Aplicar el algoritmo de expansión a partir de semillas (region growing) de manera que únicamente los lúmenes sean blancos
# y el resto de la imagen negra. Pista: utilizar dos semillas. Nota: Se pueden fijar las semillas de manera manual, pero
# se valorará positivamente a aquell@s que desarrollen una función para encontrarlas automáticamente.
# Visualizar la máscara resultante. 
mask = cleaned_mask.copy()

# Definir una imagen negra de referencia
h, w = cleaned_mask.shape

print(h, w)
ref = np.zeros((h+2, w+2), np.uint8)

#print(ref)

# Fijar la semilla. En este caso en la esquina superior izquierda (posición 0,0) para rellenar a partir de ahí
cv2.floodFill(mask, ref, (0,0), 0); # Como nos interesa rellenar de negro, utilizamos el 0 en la última posición
cv2.floodFill(mask, ref, (w-1,h-1), 0);

fig, ax = plt.subplots(1,2)
ax[0].imshow(cleaned_mask, cmap='gray'), ax[0].set_title('Máscara Otsu (Cleaned mask)')
ax[1].imshow(mask, cmap='gray'), ax[1].set_title('máscara')
plt.show()
```

#### 5) Rellenar los objetos de los lúmenes

```python
# Rellenar los lúmenes con la función binary_fill_holes de la librería scipy.ndimage.morphology
# Visualizar la máscara resultante
from scipy import ndimage

filled_image = ndimage.binary_fill_holes(mask)

plt.imshow(filled_image, cmap='gray')
plt.title('Máscara lumenes rellenos')
plt.axis('off')
plt.show()
```

#### 6) Detectar y dibujar los contornos de los lúmenes sobre la imagen original

```python
lab, num = label(filled_image, return_num=True)

print('número de objetos: ', num)
v,c = np.unique(lab, return_counts=True)
print('posibles valores de intensidad: ', v)

plt.imshow(lab, cmap='gray')
plt.axis('off')
plt.show()
```

```python
objeto = lab == 0
objeto = objeto.astype('uint8')
plt.imshow(objeto, cmap='gray')
plt.show()
```

```python
# Dibujar los contornos de los lúmenes en color verde sobre la imagen original RGB. Nota: Utilizar los flags necesarios
# para que los contornos en verde sean perfectamente visibles. 
# Visualizar la imagen superpuesta

imagen=img_1.copy()

for i in v[1:]:
    objeto = lab == i
    objeto = objeto.astype('uint8')
    convex_image = convex_hull_image(objeto)
    convex_image = convex_image.astype('uint8')
    conts,_ = cv2.findContours(convex_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) # Encontramos los contornos en una máscara 
    imagen = cv2.drawContours(imagen, conts, -1, (0,255,0), 5)
    
plt.imshow(imagen, cmap='gray')
plt.title('Smagen superpuesta')
plt.axis('off')
plt.show()
```

#### 7) Identificar y cropear el lumen más grande

```python
# Determinar cuál es el lumen de mayor área y hacer un crop del mismo sobre la imagen original RGB.
# Visualizar el lumen cropeado.

new_img = imagen[400:600,500:770]
figs, ax = plt.subplots(1,2)
ax[0].imshow(imagen, cmap='gray'), ax[0].set_title('original')
ax[1].imshow(new_img, cmap='gray'), ax[1].set_title('crop')
plt.show()
```

#### 8) Extraer 13 características geométricas que permitan caracterizar el lumen recortado 

```python
# Calcular las siguientes características del crop del lumen de mayor área, redondeando su valor hasta el cuarto decimal.
# 1) Área
# 2) Área de la bounding box
# 3) Área convexa
# 4) Exentricidad
# 5) Diámetro equivalente
# 6) Extensión
# 7) Diámetro Feret
# 8) Longitud del eje mayor
# 9) Longitud del eje menor
# 10) Orientación
# 11) Perímetro
# 12) Solidez
# 13) Compacidad
objeto = lab == 23
objeto = objeto.astype('uint8')
plt.imshow(objeto, cmap='gray')
plt.show()


A,BB,CA,E,ED,EX,MA,MiA,OR,P,S,CO,R , DF= ['area'], ['bbox_area'], ['convex_area'], ['eccentricity'], ['equiv_diameter'], \
['extent'], ['major_axis'], ['minor_axis'], ['orientation'], ['perimeter'], ['solidity'], ['compactness'], ['rectangularity'], ['Feret Diameter']

objeto = lab == 23
prop = regionprops(objeto.astype(np.uint8))

region = prop[0]

# Obtener la matriz de coordenadas de la región
coords = region.coords
    
A.append(np.round(prop[0].area, 4))
BB.append(np.round(prop[0].bbox_area, 4))
CA.append(np.round(prop[0].convex_area, 4))
E.append(np.round(prop[0].eccentricity, 4))
ED.append(np.round(prop[0].equivalent_diameter, 4))
EX.append(np.round(prop[0].extent, 4))
MA.append(np.round(prop[0].major_axis_length, 4))
MiA.append(np.round(prop[0].minor_axis_length, 4))
OR.append(np.round(prop[0].orientation, 4))
P.append(np.round(prop[0].perimeter, 4))
S.append(np.round(prop[0].solidity, 4))
CO.append(np.round(4*math.pi*prop[0].area/prop[0].perimeter**2, 4))
R.append(np.round(prop[0].area/prop[0].bbox_area, 4))
DF.append(np.max(np.linalg.norm(coords - region.centroid, axis=1)) * 2)

my_data = [tuple(A), tuple(BB), tuple(CA), tuple(E), tuple(ED), tuple(EX), \
          tuple(MA), tuple(MiA), tuple(OR), tuple(P), tuple(S), tuple(CO), tuple(R), tuple(DF)]

print(tabulate(my_data))
```
